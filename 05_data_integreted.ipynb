# Data Merge

```python
import pandas as pd

data1 = {
    'id' : [1,2,3],
    'Name': ['Ahmad', 'Mihsin', 'Zozo'],
    'age': [23,34,65]

}
data2 = {
    'id' : [1,2,3],
    'City': ['Karchi', 'Multan', 'lahore'],
    'Job': ['Doc' , 'IT' , 'Polis']

}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merg_data = pd.merge(df1 , df2 , on='id' , how='inner')

merg_data.head()
```

# Concatenation

```python
import pandas as pd

data1 = {
    'id': [1, 2, 3],
    'Name': ['Ahmad', 'Mihsin', 'Zozo'],
    'age': [23, 34, 65]
}

data2 = {
    'id': [1, 2, 3],
    'City': ['Karchi', 'Multan', 'Lahore'],
    'Job': ['Doc', 'IT', 'Polis']
}

# Convert dictionaries to DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate DataFrames along the rows (axis=0)
merged = pd.concat([df1, df2], axis=1)

# Display the concatenated DataFrame
print(merged)
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate DataFrames along the rows (axis=0)
merged = pd.concat([df1, df2], axis=0)

# Display the concatenated DataFrame
print(merged)
```

# Standarization / Scaling

```python
import pandas as pd

from sklearn.preprocessing import StandardScaler, MinMaxScaler , MaxAbsScaler


df = {
    'age' : [25,30,35,40,45],
    'height': [165,170,175,180,185],
    'weight': [55,60,65,70,75]
}

df = pd.DataFrame(df)

df

```

```python
# import scaler

scaler = StandardScaler()

scaled_df = scaler.fit_transform(df)

scaled_df
```

```python
# Convert this scaled data into pandas DataFrame
# Convert the data into StandardScaler 
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
# Convert the data into MinMaxscaler

scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
# Convert the data into MaxAbscaler

scaler = MaxAbsScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

# Load the Iris dataset
iris_data = load_iris()

# Split the dataset into features (X) and target (y)
X = iris_data.data
y = iris_data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

# Initialize the k-Nearest Neighbors classifier with k=50
knn = KNeighborsClassifier(n_neighbors=50)

# Train the classifier on the training data
knn.fit(X_train, y_train)  

# Make predictions on new data
prediction = knn.predict([[4.9, 3, 1.6, 0.4]])
print("Predicted class:", iris_data.target_names[prediction[0]])
```
